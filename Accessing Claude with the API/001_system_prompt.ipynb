```python
# load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# create an API agent
from anthropic import Anthropic

client = Anthropic(base_url="https://www.xiaoyaoapi.com")

# model = "claude-opus-4-8"
model = "claude-fable-5"
```

```python
# create the functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system_prompt=None):
    param = {"model": model, "max_tokens": 1000, "messages": messages}
    if system_prompt:
        param["system"] = system_prompt
    message = client.messages.create(**param)
    return "".join([block.text for block in message.content if block.type == "text"])
```

```python
from IPython.display import Markdown, display

messages = []
system_prompt = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """
add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")
answer = chat(messages, system_prompt=system_prompt)

display(Markdown(answer))
```
