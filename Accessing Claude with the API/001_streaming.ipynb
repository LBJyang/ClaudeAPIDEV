```python
# create an API client
from anthropic import Anthropic

client = Anthropic(base_url="https://www.xiaoyaoapi.com")
model = "claude-fable-5"
```

```python
# create the functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


# def chat(messages, system_prompt=None):
#     param = {
#         "model": model,
#         "max_tokens": 1000,
#         "messages": messages,
#     }
#     if system_prompt:
#         param["system"] = system_prompt
#     message = client.messages.create(**param)
#     return "".join([block.text for block in message.content if block.type == "text"])
```

```python
messages = []
add_user_message(messages, "给我讲个小故事。")
with client.messages.stream(model=model, max_tokens=1000, messages=messages) as stream:
    for text in stream.text_stream:
        print(text, end="")
```
